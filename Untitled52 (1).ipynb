# **Convert date columns and create additional features**

```python
df['ordered_place_time']=pd.to_datetime(df['ordered_place_time'],dayfirst=True)
df['orederd_deliver_time']=pd.to_datetime(df['ordered_deliver_time'],dayfirst=True)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['qty'] = pd.to_numeric(df['qty'], errors='coerce')
```

```python
df.dtypes
```

```python
df['day']=df['ordered_place_time'].dt.day
df['hour']=df['ordered_place_time'].dt.hour
df['weekday']=df['ordered_place_time'].dt.day_name()
```

## **Q.1 Hourly demand analysis (Most important for forcasting)**

```python
hourly_demand=df.groupby('hour')['order_id'].count()
plt.figure(figsize=(10,5))
plt.fill_between(hourly_demand.index, hourly_demand.values, alpha=0.3)
plt.plot(hourly_demand.index, hourly_demand.values)
plt.title("Hourly Order Demand (Area Chart)")
plt.xlabel("Hour of Day")
plt.ylabel("Orders")
plt.xticks(range(24))
plt.tight_layout()
plt.show()
```

## **Q.2 City Level Demand Analysis**

```python
city_demand=df.groupby('store_id')['order_id'].count().sort_values(ascending=False)
print(city_demand)
```

# **Q.3 Store Performance (Demand per store)**

```python
store_demand=df.groupby('store_id')['order_id'].count().sort_values(ascending=False)
print(store_demand)
top10 = store_demand.head(10)

plt.figure(figsize=(10,5))
plt.plot(top10.index.astype(str), top10.values)
plt.title("Top 10 Stores by Order Volume")
plt.xlabel("Store ID")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## **Q.4 SKU Category Demand (Stock Keeping Unit)**

```python
category_demand = df.groupby('sku_category')['order_id'].count().sort_values(ascending=False)
print(category_demand)
plt.figure(figsize=(10,5))
plt.bar(category_demand.index.astype(str),category_demand)
plt.title("Order By SKU Category")
plt.xlabel("SKU Category")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## **Q.5 Daily Revenue Calculations**

```python
df['revenue']=df['price']*df['qty']
df['order_date']=df['ordered_place_time'].dt.date
```

```python
daily_revenue = df.groupby('order_date')['revenue'].sum()
plt.figure(figsize=(10,5))
plt.plot(daily_revenue.index, daily_revenue.values, marker='o')
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python

peak_day = daily_revenue.idxmax()
low_day = daily_revenue.idxmin()
```

```python
category_revenue = df.groupby('sku_category')['revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
plt.bar(category_revenue.index.astype(str), category_revenue.values)
plt.title("Category-wise Revenue")
plt.xlabel("SKU Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **Q.6 Cancellation Insights**

```python
df['is_cancelled'] = df['status'].str.contains('cancel', case=False, na=False)
df['order_hour'] = df['ordered_place_time'].dt.hour
cancel_by_hour = df[df['is_cancelled']].groupby('order_hour')['order_id'].count()
plt.figure(figsize=(10,5))
plt.bar(cancel_by_hour.index, cancel_by_hour.values)
plt.title("Cancellations by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Cancellation Count")
plt.xticks(range(24))
plt.tight_layout()
plt.show()


```

# **Q.7 Zones with high cancellations**

```python
cancel_by_zone = df[df['is_cancelled']].groupby('zone')['order_id'].count().sort_values(ascending=False)
plt.figure(figsize=(10,5))
plt.bar(cancel_by_zone.index.astype(str), cancel_by_zone.values)
plt.title("Cancellations by Zone")
plt.xlabel("Zone")
plt.ylabel("Cancellation Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Q.8 Does delivery time cause cancellations?**

```python
cancelled = df[df['is_cancelled']]
non_cancelled = df[df['is_cancelled'] == False]
plt.figure(figsize=(7,5))
plt.boxplot([cancelled['delivery_minutes'].dropna(),
                 non_cancelled['delivery_minutes'].dropna()],
                labels=['Cancelled', 'Non-Cancelled'])
plt.title("Delivery Time Comparison")
plt.ylabel("Delivery Minutes")
plt.tight_layout()
plt.show()
```

## **Q.9 Correlation Analysis (Delivery Time, Distance, Price)**

```python
corr = df[['qty', 'price', 'delivery_minutes', 'store_rent']].corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
